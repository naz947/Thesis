```python
import numpy as np
import GPy
import matplotlib.pyplot as plt
```

```python
def Kernel1(x1,x2):
    value=np.exp(-np.linalg.norm((x1-x2)**2)/2)
    return(value)
def Kernel2(x1,x2):
    value=(((x1.T)*(x2))+2)**2
    return value
def Kernel3(x1,x2):
    value=np.exp(-np.abs(x1-x2))
    return value
def covariance(train,test,kernel):
    x=list(train)
    for i in range(len(test)):
        x.append(test[i])
    values=np.zeros((len(x),len(x)))
    for i in range(len(x)):
        for j in range(len(x)):
            values[i,j]=kernel(x[i],x[j])
    return values
```

```python
sample_size = 5
X = np.random.uniform(0, 5, (sample_size, 1))
Y = np.sin(X) + np.random.randn(sample_size, 1)*0.05
```

```python
n = 5
Xtest = np.linspace(-1, 1, n).reshape(-1,1)
y=np.sin(Xtest)
```

```python
plot1=covariance(Xtest,y,Kernel1)
```

```python
plot2=covariance(Xtest,y,Kernel2)
```

```python
plot3=covariance(Xtest,y,Kernel3)
```

```python
plt.plot(plot1)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```

```python
plt.plot(plot2)
plt.show()
```
